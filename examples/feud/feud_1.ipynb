```python
def capability(peak_age=20, aging_rate=1):
    # compress aging rate into (0.4, 2.1) range assymetrically centered around (1,1)
    aging_rate = ((tanh(aging_rate-1)+1)**1.2-1)*.9+1
    
    up_rate=peak_age/pi
    down_rate = peak_age/pi*3/aging_rate
    
    norm = 1
    def func(x):
        up = (tanh((-peak_age/2 + x)/up_rate)+1)/2
        down = (tanh((peak_age*(4-aging_rate*1.2) - x)/down_rate)+1)/2
        return up * down * norm 
    
    samples = linspace(peak_age-up_rate/3, peak_age+down_rate/3, 10)
    norm = (1-0.00001)/amax(func(samples))
    return func


plt_max = 120
plt_step = 10
plt_rng = arange(plt_max)

f1 = capability()
f2 = capability(peak_age=30)
f3 = capability(peak_age=30, aging_rate=4)

fig = figure(figsize=(10,4))

plot(f1(plt_rng), label='1')
plot(f2(plt_rng), label='2')
pl3, = plot(f3(plt_rng), label='3')

legend()
xticks(range(0, plt_max + plt_step, plt_step))
xlim(0,plt_max)
yticks(arange(0, 1.1, 0.1))
ylim(0,1.1)
tight_layout(pad=0.5)

@interact
def update(peak=20., aging=1.):
    f = capability(peak_age=peak, aging_rate=aging)
    pl3.set_ydata(f(plt_rng))
    fig.canvas.draw_idle()
    
```

```python
def lookup(xp, yp, left=None, right=None):
    from numpy import interp
    def func(x):
        return interp(x, xp, yp, left=left, right=right)
    return func
        
```
